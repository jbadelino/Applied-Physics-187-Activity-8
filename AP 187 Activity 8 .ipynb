```python
def tristimulus_vals(P, ciex, ciey, ciez):
    X = P*ciex
    Y = P*ciey
    Z = P*ciez
    wav = np.linspace(380,780, num=81)
    X_area = np.trapz(X, wav)
    Y_area = np.trapz(Y, wav)
    Z_area = np.trapz(Z, wav)
    return X_area, Y_area, Z_area
```

```python
#CIE xyz 
def cie_xyz(X, Y,Z):
    I = X + Y + Z
    x = X/I
    y = Y/I
    z = 1 - x- y
    return x,y
```

```python
def f(t):
    if t > (6/29)**3:
        return t**(1/3)
    else: 
        return (1/3)*((29/6)**2)*t + (4/29)
```

```python

def Lab_ch(x,y,z):
    L = 116*(y) - 16
    a = 500*(x-y)
    b = 200*(y-z)
    C = np.sqrt(a**2 + b**2)
    h = np.radians(np.arctan(b/a))
    return L,a,b,C,h
```

```python
def CIE76delta_E(L1,L2,a1,a2,b1,b2):
    delta_E12 = np.sqrt((L2-L1)**2 + (a2-a1)**2 + (b2-b1)**2)
    return delta_E12
```

```python
def CHdelta_E(L1,L2,C1,C2,h1,h2):
    delta_E12 = np.sqrt((L2-L1)**2 + (C2-C1)**2 + (h2-h1)**2)
    return delta_E12
```

```python
cie76_delta_E = []
ch_delta_E = []
```

```python
#Lab transformation - patch 1
patcha = tristimulus_vals(patch1_new*D65_new,xbar_new, ybar_new, zbar_new)
patchb = tristimulus_vals(patch1_new*D55_new,xbar_new, ybar_new, zbar_new)

#f(tristimulus values)
d65p = np.array(patcha)/np.array(D65_light)
d55p = np.array(patchb)/np.array(D55_light)

f_d65px = f(d65p[0])
f_d65py = f(d65p[1])
f_d65pz = f(d65p[2])

f_d55px = f(d55p[0])
f_d55py = f(d55p[1])
f_d55pz = f(d55p[2])

f_d65p_lab = Lab_ch(f_d65px, f_d65py, f_d65pz)
f_d55p_lab = Lab_ch(f_d55px, f_d55py, f_d55pz)

patch_cie1 = CIE76delta_E(f_d65p_lab[0], f_d55p_lab[0],f_d65p_lab[1], f_d55p_lab[1],f_d65p_lab[2], f_d55p_lab[2])
patch_ch1 = CHdelta_E(f_d65p_lab[0], f_d55p_lab[0],f_d65p_lab[3], f_d55p_lab[3],f_d65p_lab[4], f_d55p_lab[4])

cie76_delta_E.append(patch_cie1)
ch_delta_E.append(patch_ch1)
```
